```python
# let's start actually building out some of our code

class Player:
    def __init__(self, fname, lname, team, age, number, position, nation, value):
        self.fname = fname
        self.lname = lname
        self.team = team
        self.age = age
        self.number = number
        self.position = position
        self.nation = nation
        self.value = value
        
    def info(self):
        """
        prints a descriptive string about this player
        """
        print(f"\t{self.number}\t{self.position}\t{self.fname} {self.lname}\t{self.team}\t{self.nation}\t{self.age}\t£{self.value}")

class Team:
    def __init__(self, name, league, location, manager, mascot, budget):
        self.players = []
        self.league = league
        self.name = name
        self.mascot = mascot
        self.location = location
        self.manager = manager
        self.total_value = 0
        self.budget = budget
        
    def show(self):
        """
        prints descriptions the team's players
        """
        # loop through the players list
        print(f"{self.name}'s Lineup:")
        for player in self.players:
            player.info()
        
    def add(self, newplayer):
        """
        purchase a new player
        """
        # take an existing player (input!)
        # add them to the team
        self.players.append(newplayer)
        # increase the total_value of the team
        self.total_value = self.total_value + newplayer.value
        # decrease the budget of the team
        self.budget = self.budget - newplayer.value
            # im not gonna worry about preventing the buy if insufficient budget
        
    def remove(self):
        """
        sell a player
        """
        if not self.players:
            print('You have no players to sell.')
            return
        # if we want to sell a player, we're shown a list of our players
            # use the show method - no reason to repeat code
        self.show()
        # the user is asked for a number to sell
        print('\n')
        try:
            num = int(input("Which player would you like to sell? (give the player's number) "))
        except:
            print('Invalid player number. No player sold.')
            return
        # we then remove the associated player from our players list
        sold = False # set up a value for the sold variable such that we can use it to determine if no player was sold
        for i in range(len(self.players)):
            if self.players[i].number == num:
                sold = self.players.pop(i) # remove the player from the team
                print(f'Sold:')
                sold.info()
                # increase our budget
                self.budget += sold.value
                # decrease our total value
                self.total_value -= sold.value
                break
        if not sold:
            print('There is no player with that number on your team. No player sold.')
        
        
    def clear(self):
        """
        sell all players
        """
        if not self.players:
            print('You have no players to sell.')
            return
        # increase budget by total_value
        self.budget += self.total_value
        print(f'You have sold all of your players. You made £{self.total_value}. Your new budget is £{self.budget}.')
        # total_value to 0
        self.total_value = 0
        # remove all players from the player list
        self.players = []
        
        
    def checkout(self):
        """
        quitting method - idk what I want to do with this yet
        """
        
def createPlayer():
    print('Please give me some information about this new player:')
    while True:
        try:
            fname = input("What is the player's first name? ")
            lname = input("What is the player's last name? ")
            team = input("What is the player's real team? ")
            age = int(input("How old is the player (integer value please)? "))
            number = int(input("What is the player's number (integer value please)? "))
            position = input("What is the player's position? ")
            nation = input("Where is the player from? ")
            value = int(input("How much does this player cost (GBP integer value please)? "))
            return Player(fname, lname, team, age, number, position, nation, value)
        except:
            choice = input('You gave some bad input. Try again? [y] to try again.' )
            if choice != 'y':
                return False
    
def createTeam():
    pass
```
